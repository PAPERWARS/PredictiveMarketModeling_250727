
IMPORTING LIBRARIES AND DATASET


```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('data.csv')
```

PROCESSING DATA


*   Removing rows with missing values (as mentioned in lecture 2)
*   Making two matrices, one with all the parameters and one with price
*   Converting array from pandas library to numpy






```python
newdata=data.dropna()
newdata= newdata.reset_index(drop=True)
X=newdata.drop(columns=['ID', 'Price'])
Y=newdata['Price']
x=X.to_numpy()
y=Y.to_numpy().reshape(-1,1)
x = np.c_[np.ones(X.shape[0]), x]
```

MATRIX-BASED REGRESSION METHOD

((X^TX)^-1)X^TY

```python
xT=x.T
xTx=xT@x
xTx_inv = np.linalg.inv(xTx)
xTy=xT@y
beta=xTx_inv@xTy
y_est=x@beta
RSS=np.sum((y-y_est)**2)
MSE=np.mean((y-y_est)**2)
rootMSE=(MSE**0.5)
avg_price=newdata['Price'].mean()
```

OUTPUT

```python
print("the RSS value is ",float(RSS))
print("The MSE value is ",float(MSE))
print("The average price of the houses is",int(avg_price))
print("The RMSE value of the model is ",int(rootMSE))






plt.scatter(y, y_est)
plt.plot(
    [y.min(), y.max()],
    [y.min(), y.max()]
)

plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Matrix Regression Model Predicted Prices")
plt.show()
```
